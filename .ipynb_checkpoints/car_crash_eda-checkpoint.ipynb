## Car Crashes in Montgomery County EDA

### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

pd.set_option('display.max_columns', 50)
```

### Import Data

```python
# Original crash data
crash_orig_df = pd.read_csv('./data/crash_reporting_drivers_data.csv', low_memory=False)
```

```python
# Modified crash data
crash_mod_df = pd.read_csv('./data/crash_data_modified.csv', low_memory=False)
```

```python
# Categorized crash data
crash_cats_df = pd.read_csv('./data/crash_categorical.csv', low_memory=False)
```

### Percentage of Drivers at-Fault vs not at-Fault

```python
crash_cats_df['Driver At Fault'].value_counts(normalize=True)
```

### Counts: Driver at-fault, Driver Drug Abuse, and Driver Distracted

```python
fig = plt.figure(figsize = (15,5))

ax1 = fig.add_subplot(1,3,1)
sns.countplot(data = crash_mod_df, x = 'Driver At Fault', ax=ax1)

ax2 = fig.add_subplot(1,3,2)
sns.countplot(data = crash_mod_df, x = 'Driver Substance Abuse', ax=ax2)

ax3 = fig.add_subplot(1,3,3)
sns.countplot(data = crash_mod_df, x = 'Driver Distracted By', ax=ax3);
```

### Count: Time of Day Crash Occurred

```python
plt.figure(figsize = (15,5))
sns.countplot(data = crash_mod_df, x = 'Crash Time of Day');
```

### Count: Surface Conditions at Time of Crash

```python
plt.figure(figsize = (15,5))
plt.xticks(rotation = 80)
sns.countplot(data = crash_mod_df, x = 'Surface Condition');
```
